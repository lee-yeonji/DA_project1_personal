### 변수 매핑
- 국민여가활동조사: `df` + `code_df`

##### 국민여가활동조사 데이터 매핑

1. "제목" 번역 실행
- `rename` 사용

```python
# '제목' 번역하기 (컬럼명 매핑)
# code_df에서 {'변수명':'항목명'} 으로 된 '컬럼 번역 규칙'을 만듦
code_dict = code_df.drop_duplicates(subset=['변수명']).set_index('변수명')['항목명'].to_dict()

# df를 복사해서 '번역본'을 만듦
df_copied = df.copy()

# df.rename() 함수를 사용해 '번역본'의 제목(컬럼) 번역
df_copied = df_copied.rename(columns=code_dict)

df_copied.head()
```

2. "내용" 번역하기 (모든 코드값을 살리는 단계)
- df_copied 내부 번역

```python
# df_copied '내용' 번역하기 (데이터 값 매핑)
# 1. 'code_df'에서 '코드값'이 있는(번역이 필요한) 변수 목록을 뽑음
code_var_list = code_df[code_df['코드값'].notna()]['변수명'].unique()

# 2. 번역할 항목('EUP', 'Q1_1' 등)을 하나씩 돌아가며 작업 진행
for var_code in code_var_list:
  col_name = code_dict.get(var_code)

  # 3. '번역본'에 없는 컬럼이면(데이터에 빠졌으면) 통과
  if not col_name or col_name not in df_copied.columns:
    continue

  # 4. "내용 번역 사전" 만들기
  # Apply the filter before setting the index
  value_map = code_df[code_df['변수명'] == var_code].set_index(code_df[code_df['변수명'] == var_code]['코드값'].astype(str))['코드명'].to_dict()

  # 5. 번역 규칙 완성: 예){'1': '시부(동)', '2': '군부(읍면)'}
  if not value_map:
    continue

  # 6. '번역본'의 컬럼들의 '내용'을 번역
  original_col_as_str = df_copied[col_name].astype(str).str.replace(r'\.0$', '', regex=True)
  df_copied[col_name] = original_col_as_str.map(value_map).fillna(original_col_as_str)
```

3. 번역본(df_copied) 파일 새로 저장하기
=> `leisure_activities.csv` 로 저장

```python
# 1. 저장할 경로와 파일명을 변수로 만듦.
output_path = "/content/gdrive/MyDrive/데이터분석 학습 및 실습/공모전-2025문화체육관광통계데이터/leisure_activities.csv"

# 2. 새로운 파일로 저장
df_copied.to_csv(output_path, index=False, encoding='utf-8-sig')
```

##### 국민문화예술활동조사 데이터 매핑


1. "제목" 번역 사전 만들기
- `column_map` 사용
- '항목명'이 중복될 경우에 발생할 오류 방지

```python
# '제목' 번역하기 (컬럼명 매핑)
# 1. code_df에서 '변수명'과 '항목명'의 1:1 관계를 만들기
# (제목 번역에만 쓸 임시 사본이기 때문에 drop_duplocates 사용)
code_ca_map = code_ca_df.drop_duplicates(subset=['변수명'])

# 2. '항목명'이 몇 번 쓰였는지 세기 (중복 여부 확인용)
name_counts = code_ca_map['항목명'].value_counts()

# 3. 'column_map' 딕셔너리를 만듦
column_map = {}
for index, row in code_ca_map.iterrows():
  var = row['변수명']
  name = row['항목명']

  # 4. 만약 '항목명'이 1번 초과로 사용되었다면 (=중복이라면)
  if name_counts[name] > 1:
    # 4-1. '항목명 (변수명)' 형태로 고유한 이름을 만듦
    unique_name = f'{name} ({var})'
  else:
    # 4-2. 중복이 아니라면, 원래 '항목명'을 그대로 사용
    unique_name = name
  column_map[var] = unique_name
```

2. "제목" 번역하기 (안전한 `rename` 사용)

```python
# 1. df_ca를 복사해서 '번역본(df_ca_copied)'를 만들기
df_ca_copied = df_ca.copy()

# 2. df.rename() 함수를 사용해서 제목 번역
df_ca_copied = df_ca_copied.rename(columns=column_map)
```

3. "내용" 번역하기 (모든 코드값을 살리는 단계)
- `df_ca_copied` 내부 번역

```python
# df_ca_copied '내용' 번역하기 (데이터 값 매핑)
# 1. 'code_ca_df'에서 '코드값'이 있는(번역이 필요한) 변수 목록을 뽑음
code_var_list2 = code_ca_df[code_ca_df['코드값'].notna()]['변수명'].unique()

# 2. 원본 '변수명'을 하나씩 순회
for var_code2 in code_var_list2:

  # 3. '변수명'에 해당하는 '고유 한글 제목'을 사전(column_map)에서 찾음
  col_name2 = column_map.get(var_code2)

  # 4. '번역본'에 해당 컬럼이 없으면(데이터에 빠졌으면 통과)
  if not col_name2 or col_name2 not in df_ca_copied.columns:
    continue

  # 4. 모든 코드값 살려서 "내용 번역 사전" 만들기
  # Apply the filter before setting the index
  value_map2 = code_ca_df[code_ca_df['변수명'] == var_code2].set_index(code_ca_df[code_ca_df['변수명'] == var_code2]['코드값'].astype(str))['코드명'].to_dict()

  if not value_map2:
    continue

  # 6. '번역본'의 컬럼들의 '내용'을 번역
  original2_col_as_str = df_ca_copied[col_name2].astype(str).str.replace(r'\.0$', '', regex=True)
  df_ca_copied[col_name2] = original2_col_as_str.map(value_map2).fillna(original2_col_as_str)
```

3. 번역본(df_ca_copied) 파일 새로 저장하기
=> `culture_activities.csv` 로 저장

```python
# Google Drive '내 드라이브' 안 폴더에 경로 만들기
output_path2 = "/content/gdrive/MyDrive/데이터분석 학습 및 실습/공모전-2025문화체육관광통계데이터/culture_activities.csv"

df_ca_copied.to_csv(output_path2, index=False, encoding='utf-8-sig')
```
